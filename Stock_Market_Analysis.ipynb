# Getting Started with Stock market Crash Analysis w Python

```python
import pandas as pd
import numpy as np

df = pd.read_csv('cleaned_sensex.csv')
print(df.head())
```

```python
if 'Date' in df.columns:
  df['Date'] = pd.to_datetime(df['Date'])
  df = df.sort_values('Date')
  df.set_index('Date', inplace=True)

print("\nDataFrame after converting 'Date' to datetime and sorting:")
print(df.head())
```

```python
df['Daily_Return'] = df['Close'].pct_change() * 100

# Define a treshold for a daily crash (e.g., drop more than 5%)
crash_threshold_daily = -5
df['Crash_Daily']  = df['Daily_Return'] <=crash_threshold_daily
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Scatter(x=df.index, y=df['Close'], mode='lines', name='Close Price'))

crash_days = df.index[df['Crash_Daily']]
crash_closes = df['Close'][df['Crash_Daily']]

fig.add_trace(go.Scatter(x=crash_days, y=crash_closes, mode='markers', marker=dict(color='red', size=10), name='Daily Crashes (<= {crash_threshold_daily}%)'))

fig.update_layout(
    title='Sensex Close Prices with Daily Crashes',
    xaxis_title='Date',
    yaxis_title='Close Price',
    xaxis_rangeslider_visible=True,
    template="plotly_white"
    )

fig.show()

fig.show()
```
